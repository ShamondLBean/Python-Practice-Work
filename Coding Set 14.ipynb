```python
#I will start with PE 2 Phonebook database
#First I will start by importing sqlite
import sqlite3
#Next I will create te phonebook database
def create_phonebook_database():
    conn = sqlite3.connect('phonebook.db')
    c = conn.cursor()
    c.execute("Create table if not exists Entries (name text primary key, phone_number text)")
    conn.commit()
    conn.close()
#Next I will create the funtion to add entries
def add_entry(name, phone_number):
    conn = sqlite3.connect('phonebook.db')
    c = conn.cursor()
    c.execute('insert into entries values (?, ?)', (name, phone_number))
    conn.commit()
    conn.close()
#Next I will create the function to look up entries
def lookup_entry(name):
    conn = sqlite3.connect('phonebook.db')
    c = conn.cursor()
    c.execute('Select phone_number from entries where name=?', (name,))
    result = c.fetchone()
    conn.close
    if result:
        return result[0]
    else:
        return 'Entry not found'
#Next I will create the function to update contacts
def update_entry(name, new_phone_number):
    conn = sqlite3.connect('phonebook.db')
    c = conn.cursor()
    c.execute('Update entries set phone_number=? where name=?', (new_phone_number, name))
    conn.commit()
    conn.close()
#Next I will create the function to delete contacts
def delete_entry(name):
    conn = sqlite3.connect('phonebook.db')
    c = conn.cursor()
    c.execute('Delete from entries where name=?', (name,))
    conn.commit()
    conn.close()
#Now I will run the CRUD application
def main():
    create_phonebook_database()
#First I will print the phonebook menu
    while True:
        print('Phonebook Menu:')
        print('1. Add a contact')
        print('2. Look up a phone number')
        print('3. Update a phone number')
        print('4. Delete a contact')
        print('5. Exit the menu')
#Next I will create the input for user choice
        choice = input('Enter Your Choice: ')
#Next I will list all of the choices
        if choice == '1':
            name = input('Enter Name: ')
            phone_number = input('Enter phone number: ')
            add_entry(name, phone_number)
            print('Contact successfully added!')
        elif choice == '2':
            name = input('Enter name to look up: ')
            result = lookup_entry(name)
            print(f'Phone Number: {result}')
        elif choice == '3':
            name = input('Enter the name to update: ')
            new_phone_number = input('Enter the new phone number: ')
            update_entry(name, new_phone_number)
            print('Contact successfully updated!')
        elif choice == '4':
            name = input('Enter the name to delete: ')
            delete_entry(name)
            print('The contact has been deleted.')
        elif choice == '5':
            print('Exiting')
            break
#Next I will create an else function in case the user types a bad number
        else:
            print('Invalid choice, Please only enter a number 1-5.')
#Lastly I will call main
if __name__ == '__main__':
    main()
                  
                
```
