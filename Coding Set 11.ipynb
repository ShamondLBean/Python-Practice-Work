```python
#PE Q.2 Supervisor class
#First I will start by creating the employee class
class Employee:
#Next I will initialize the attributes
    def __init__(self, employee_name, employee_number):
        self.__employee_name = employee_name
        self.__employee_number = employee_number
#Next I will create the settor methods
    def set_employee_name(self, employee_name):
        self.__employee_name = employee_name
    def set_employee_number(self, employee_number):
        self.__employee_number = employee_number
#Next I will create the gettor methods
    def get_employee_name(self):
        return self.__employee_name
    def get_employee_number(self):
        return self.__employee_number
#Next I will create a class for a production worker
class ProductionWorker(Employee):
    def __init__(self, employee_name, employee_number, shift_number, hourly_pay):
#Next I will call the superclass init
        Employee.__init__(self, employee_name, employee_number)
#Next I will initialize the production attributes
        self.__shift_number = shift_number
        self.__hourly_payrate = hourly_pay
#Next I will use the settor methods
    def set_shift_number(self, shift_number):
        self.__shift_number = shift_number
    def set_hourly_payrate(self, hourly_pay):
        self.__hourly_payrate = hourly_pay
#Next I will use the gettor methods
    def get_shift_number(self):
        return self.__shift_number
    def get_hourly_payrate(self):
        return self.__hourly_payrate
#Next I will create a supervisor class
class Supervisor(Employee):
#Next I will initialize the supervisor attributes
    def __init__(self, employee_name, employee_number, annual_salary, annual_bonus):
#Next I will call the employee class again
        Employee.__init__(self, employee_name, employee_number)
#Next I will initialize the salary and bonus attributes
        self.__annual_salary = annual_salary
        self.__annual_bonus = annual_bonus
#Next I will use the settor method
    def set_annual_salary(self, annual_salary):
        self.__annual_salary = annual_salary
    def set_annual_bonus(self, annual_bonus):
        self.__annual_bonus = annual_bonus
#Next I will use the gettor method
    def get_annual_salary(self):
        return self.__annual_salary
    def get_annual_bonus(self):
        return self.__annual_bonus
```

```python
#PE Q.3 Person and Customer
#I will start with the person class
class Person:
#Next I will initialize the attributes
    def __init__(self, name, address, number):
        self.__name = name
        self.__address = address
        self.__number = number
#Next I will create the settor methods
    def set_name(self, name):
        self.__name = name
    def set_address(self, address):
        self.__address = address
    def set_number(self, number):
        self.__number = number
#Next I will create the Gettor methods
    def get_name(self):
        return self.__name
    def get_address(self):
        return self.__address
    def get_number(self):
        return self.__number
#Next I will create the subclass for the customer
class Customer(Person):
#Next I will initialize the attributes 
    def __init__(self, name, address, number, would_like_to_join_mailing_list = False):
        Person.__init__(self, name, address, number)
        self.__would_like_to_join_mailing_list = would_like_to_join_mailing_list
#Next I will create the settor for if the customer would like to join the mailing list
    def set_wishes(self, would_like_to_join_mailing_list):
        self.__would_like_to_join_mailing_list = would_like_to_join_mailing_list
#Next I will create the gettor for this as well
    def get_wishes(self):
        return self.__would_like_to_join_mailing_list
#Next I will call the main function to print the customers info and use a if else structure for the mailing list
def main():
#Next I will create the profile of the customer
    customer = Customer('Shamond Bean', 'Bean@gmail.com', '269-092-9821')
#Next I will print the information of the customer
    print('Customer Name:', customer.get_name())
    print('Customer Address:', customer.get_address())
    print('Customer Number:', customer.get_number())
#Next I will use the if statement for if the customer would like to be on the mail list
    if customer.set_wishes():
        print('You would like to join our mailing list.')
#Next I will use the else statement if they would not like to be on the mail list
    else:
        print('You will not be added to our mailing list.')
#Lastly I will call the main function again
main()
```
